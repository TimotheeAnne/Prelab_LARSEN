```python
import cma
import numpy as np 
import matplotlib.pyplot as plt 
import pickle 
```

```python
def f(x):
    return np.linalg.norm(x-np.array([0.9,0.2]))
```

```python
es = cma.CMAEvolutionStrategy(2 * [0], 0.3, { 'bounds': [-np.ones(2), np.ones(2)*0.9]})
while not es.stop():
    solutions = es.ask()
    es.tell(solutions, [f(x) for x in solutions])
    es.logger.add()  # write data to disc to be plotted
    es.disp()
    print(es.result[6])
```
